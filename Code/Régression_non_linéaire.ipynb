#Régression non linéaire
Ce Notebook traite de la création d'un jeu de données bruités non linéaires et de la création d'un modèle adapté à ce problème

##Dépendances

```python
import numpy as np
import numpy.random as rd
import math
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from google.colab import files
```

## Création du jeu de données

```python
x = rd.rand(600)*10
y = np.array([0.1 * math.cos(i) * i + rd.rand() * 0.2 for i in x])

plt.scatter(x[::1], y[::1], s=2)
plt.grid()
plt.title("Données à modéliser")
plt.show()
plt.savefig('dataset.png',dpi=300)
```

## Création du modèle et execution du modèle

```python
model = tf.keras.Sequential()
model.add(Dense(64,activation = "relu",input_dim = 1))
model.add(Dense(64,activation = "relu"))
model.add(Dense(1,activation = "linear"))

model.compile(loss = "mse", optimizer = "adam")
model.summary()
```

```python
model.fit(x, y, epochs=300)
```

```python
abc = np.linspace(0,10,num = 2000)
y_predicted = model.predict(abc)

# Display the result
plt.scatter(x[::1], y[::1], s=1)
plt.plot(abc, y_predicted,color = "r")
plt.grid()
plt.show()
```

##On essaye avec une autre fonction non linéaire


```python
x = np.concatenate((rd.rand(300)* -20,rd.rand(300)*20),axis = 0)
y = np.array([15 * np.cos(i/3) - 12 * np.sin(i)  + (rd.rand() * 10) for i in x])
print(np.size(x))
print(np.size(y))
plt.scatter(x, y)
plt.grid()
plt.title("Données à modéliser")
plt.show()
```

```python
# Display the result
abc = np.linspace(-20,20,num = 2000)
y_predicted = model.predict(abc)

plt.scatter(x[::1], y[::1], s=1)
plt.plot(abc, y_predicted,color = "r")
plt.grid()
plt.show()
```

## On crée un réseau à 4 couches pour rendre comptes de dépendances en x^4

```python
model = tf.keras.Sequential()
model.add(Dense(256,activation = "relu",input_dim = 1))
model.add(Dense(64,activation = "relu"))
model.add(Dense(64,activation = "relu"))
model.add(Dense(256,activation = "relu"))
model.add(Dense(1,activation = "linear"))

model.compile(loss = "mse", optimizer = "adam")
model.summary()
```

```python
model.compile(loss = "mse", optimizer = "adam")
model.summary()
model.fit(x, y, epochs=1000)
```

```python
# Display the result
abc = np.linspace(-20,20,num = 2000)
y_predicted = model.predict(abc)

plt.scatter(x[::1], y[::1], s=1)
plt.plot(abc, y_predicted,color = "r")
plt.grid()
plt.show()
```
